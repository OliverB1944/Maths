```python
import numpy as np
import matplotlib.pyplot as plt
import math #for cos and sin and pi

def rotationY(x, y, z, angle) :
  angle = degreeToRadians(angle)
  zAngle = z*(math.cos(angle)) - x*(math.sin(angle))
  xAngle = z*(math.sin(angle)) + x*(math.cos(angle))
  yAngle = y
  return xAngle, yAngle, zAngle

def rotationZ(x, y, z, angle) :
  angle = degreeToRadians(angle)
  xAngle = x*(math.cos(angle)) - y*(math.sin(angle))
  yAngle = x*(math.sin(angle)) + y*(math.cos(angle))
  zAngle = z
  return xAngle, yAngle, zAngle

def rotationX(x, y, z, angle) :
  angle = degreeToRadians(angle)
  yAngle = y*(math.cos(angle)) - z*(math.sin(angle))
  zAngle = y*(math.sin(angle)) + z*(math.cos(angle))
  xAngle = x
  return xAngle, yAngle, zAngle

def degreeToRadians(angle) :
  radians = angle * (math.pi/180)
  return radians

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

coordinates1 = rotationX(-10,0,10, -30)
coordinates2 = rotationY(-10,0,10, 175)
coordinates3 = rotationZ(-10,0,10, 65)

ax.scatter(coordinates1[0], coordinates1[1], coordinates1[2])
ax.scatter(coordinates2[0], coordinates2[1], coordinates2[2])
ax.scatter(coordinates3[0], coordinates3[1], coordinates3[2])
```

```python
import matplotlib.pyplot as plt

#circle parameters
radius = 1
center = (0, 0)

fig, ax = plt.subplots()

#plots the circle
circle = plt.Circle(center, radius, color='blue', fill=False)

#plots circle
ax.add_artist(circle)

#set aspect ration so it doesnt look like an oval
ax.set_aspect('equal')

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#creates 100 points ranging from 0 to 2pi for u
u = np.linspace(0, 2 * np.pi, 100)
#creates 100 points ranging from 0 to pi for v
v = np.linspace(0, np.pi, 100)
#coordinates for sphere
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))

#plots sphere
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='b')

#set equal aspect ration so that plots are consistent across xis
ax.set_box_aspect([1,1,1])
```
